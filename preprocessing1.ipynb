Pembersihan Teks

```python
sebelum_tahap3 = df.loc[sampel_indeks, 'clean_text'].copy()
```

```python
def clean_tahap3(text):
    text = str(text)
    # Menghapus mention (@username)
    text = re.sub(r'\@\w+', '', text)
    # Menghapus simbol hashtag (#)
    text = re.sub(r'#', '', text)
    # Menghapus semua karakter selain huruf abjad (a-z) dan spasi
    # (Ini otomatis akan menghapus emoji, tanda baca, dan angka)
    text = re.sub(r'[^a-z\s]', ' ', text)
    # Menghapus spasi ganda atau berlebih dan merapikan ujung teks
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df['clean_text'] = df['clean_text'].apply(clean_tahap3)
```

```python
sesudah_tahap3 = df.loc[sampel_indeks, 'clean_text'].copy()
```

```python
tabel_perbandingan_tahap3 = pd.DataFrame({
    'Sebelum': sebelum_tahap3,
    'Sesudah': sesudah_tahap3
})
```

```python
print("Penghapusan emoji, mention, dan karakter")
pd.set_option('display.max_colwidth', None)
display(tabel_perbandingan_tahap3)
```

NORMALISASI TEKS

```python
sebelum_tahap4 = df.loc[sampel_indeks, 'clean_text'].copy()
```

```python
url_kamus = 'https://raw.githubusercontent.com/nasalsabila/kamus-alay/master/colloquial-indonesian-lexicon.csv'
```

```python
df_kamus = pd.read_csv(url_kamus)
```

```python
slang_dict_publik = dict(zip(df_kamus['slang'], df_kamus['formal']))
```

```python
kamus_kustom = {
    'amp': 'dan', 
    'kalo': 'kalau', 
    'bgt': 'banget', 
    'bngt': 'banget',
    'siiiikkkkkkkk': 'sih', 
    'mlarat': 'melarat', 
    'jijiq': 'jijik',
    'wis wayahe': 'sudah waktunya', 
    'saatx': 'saat', 
    'bangettt': 'banget',
    'kebalikaaaannn': 'kebalikan',
    'bpk': 'bapak',
    'bangkruuut': 'bangkrut',
    'buzzerp': 'buzzer',
    'benaran': 'beneran',
    'gelo': 'gila',
    'Kagok': 'tanggung',
    'puyeng': 'pusing',
    'ink': 'ini',
    'dibuvarkan': 'dibubarkan',
    'insfrastruktur': 'infrastruktur',
    'trilyun': 'triliun',
    'syulit': 'sulit',
    'ikaen': 'ikn',
    'yaelah': 'waduh',
    'maksa': 'memaksa',
    'gpp': 'tidak apa-apa',
    'tau': 'tahu',
    'diih': 'jijik',
    'now': 'sekarang',
    'kebagian': 'mendapat bagian',
    'bree': 'bro',
    'wamen': 'wakil menteri',
    'aja': 'saja',
    'jokodok': 'jokowi',
    'hater': 'pembenci'
    
}
slang_dict_publik.update(kamus_kustom)
```

```python
def normalisasi_teks(text):
    text = str(text)
    kata_kata = text.split()
    # Mencocokkan dengan kamus yang sudah digabung
    kata_normal = [slang_dict_publik.get(kata, kata) for kata in kata_kata]
    return ' '.join(kata_normal)
```

```python
df['clean_text'] = df['clean_text'].apply(normalisasi_teks)
```

```python
sesudah_tahap4 = df.loc[sampel_indeks, 'clean_text'].copy()
```

```python
tabel_perbandingan_tahap4 = pd.DataFrame({
    'Sebelum': sebelum_tahap4,
    'Sesudah': sesudah_tahap4
})
```

```python
print("NORMALISASI TEKS")
pd.set_option('display.max_colwidth', None)
display(tabel_perbandingan_tahap4)
```

Memfilter data (hanya bahasa Indonesia, hapus baris kosong, dan duplikat)

```python
print(f"Jumlah baris sebelum difilter: {len(df)}")
```

```python
df['clean_text'] = df['clean_text'].replace(r'^\s*$', pd.NA, regex=True)
```

```python
df = df.dropna(subset=['clean_text'])
```

```python
df = df.drop_duplicates(subset=['clean_text'], keep='first')
```

```python
df = df[df['clean_text'].astype(str).str.split().str.len() >= 3]
```

```python
DetectorFactory.seed = 0

def filter_bahasa_indonesia(text):
    try:
        return detect(str(text)) == 'id'
    except LangDetectException:
        return False

mask_indonesia = df['clean_text'].apply(filter_bahasa_indonesia)

df = df[mask_indonesia].copy()

df = df.reset_index(drop=True)
```

```python
print(f"Jumlah baris setelah difilter: {len(df)}")
```

```python
lokasi_simpan = 'dataset_twitter_clean_final.csv'

df[['clean_text']].to_csv(lokasi_simpan, index=False)

print(f"lokasi: {lokasi_simpan}")
```
